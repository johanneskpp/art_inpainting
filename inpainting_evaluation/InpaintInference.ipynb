# 02 inference with mmagic default inpainting models
## a) import dependencies

```python
import mmcv
import matplotlib.pyplot as plt
from mmagic.apis import MMagicInferencer
from PIL import Image
import os
import glob
```

## b) create inferencer

You can find all available models/methods for inpainting here: https://github.com/open-mmlab/mmagic

```python
model_name = "aot_gan"
editor = MMagicInferencer(model_name, model_setting=0)
```

## c) iterate over folder and store inference results

```python
i = 0

file_list = glob.glob("/content/drive/MyDrive/TUD/defectsegmentation/dataset_small/img/test/*.png")
processed = glob.glob("/content/drive/MyDrive/TUD/defectsegmentation/dataset_small/res_{}/test/*.png".format(model_name))
```

```python
for img_file in file_list:
  msk_file = img_file.replace("/img/", "/msk/")
  res_file = img_file.replace("/img/", "/res_{}/".format(model_name))

  if not res_file in processed:

    image = editor.infer(text="abc", img=img_file, mask=msk_file, height=256, width=256, result_out_dir=res_file)

    i += 1
  else:
    print("skipping {} cause already processed".format(img_file))
  print(i, end="\r")
```

# 03 inference with stable diffusion 1.5 for inpainting and empty prompt
## a) init stable diffuser

```python
from mmengine import MODELS, Config
from mmengine.registry import init_default_scope

init_default_scope('mmagic')

config = '/content/stable-diffusion_ddim_denoisingunet-inpaint.py'
config = Config.fromfile(config).copy()

StableDiffuser = MODELS.build(config.model)
prompt = ''
StableDiffuser = StableDiffuser.to('cuda')
```

## b) define paths

```python
i = 0
file_list = glob.glob("/content/drive/MyDrive/TUD/defectsegmentation/dataset_small/img/test/*.png")
processed = glob.glob("/content/drive/MyDrive/TUD/defectsegmentation/dataset_small/res_sd15/test/*.png")
```

## c) run inference

```python
for img_file in file_list:
  msk_file = img_file.replace("/img/", "/msk/")
  res_file = img_file.replace("/img/", "/res_sd15/")

  if not res_file in processed:
    image = Image.fromarray(mmcv.imread(img_file, channel_order='rgb'))
    mask = Image.fromarray(mmcv.imread(msk_file)*255).convert('L')


    image = StableDiffuser.infer(
        prompt,
        image,
        mask,
        num_inference_steps = 100
    )['samples'][0]
    image.save(res_file)

    i += 1
  else:
    print("skipping {} cause already processed".format(img_file))
  print(i, end="\r")
```
