# 02 define augmentation transforms for masks

```python
transform = A.Compose(
    [
        A.CoarseDropout(p=0.4, max_holes=20, max_height=8, max_width=8, min_holes=1, min_height=8, min_width=8, fill_value=(255, 255, 255), mask_fill_value=None),
        A.VerticalFlip(p=0.3),
        A.HorizontalFlip(p=0.3),
        A.GridDistortion(p=0.3, num_steps=5, distort_limit=(-0.3, 0.3), interpolation=0, border_mode=0, value=(255, 255, 255), mask_value=None, normalized=False),
        A.Blur(blur_limit=(15, 15), p=0.3),
        A.Defocus(p=0.3, radius=(3, 10), alias_blur=(0.1, 0.5)),
        A.Downscale(p=0.3, scale_min=0.25, scale_max=0.25),
        A.MotionBlur(p=0.4, blur_limit=(5, 9), allow_shifted=True),
        A.Perspective(p=0.3, scale=(0.25, 0.58), keep_size=1, pad_mode=0, pad_val=(0, 0, 0), mask_pad_val=0, fit_output=0, interpolation=0),
        A.Rotate(p=0.3, limit=(-90, 90), interpolation=0, border_mode=0, value=(0, 0, 0), mask_value=None, rotate_method='largest_box', crop_border=False),
        A.ShiftScaleRotate(p=0.5, shift_limit_x=(-0.06, 0.06), shift_limit_y=(-0.06, 0.06), scale_limit=(-0.09999999999999998, 0.10000000000000009), rotate_limit=(-90, 90), interpolation=0, border_mode=0, value=(0, 0, 0), mask_value=None, rotate_method='largest_box'),
        A.ZoomBlur(p=0.5, max_factor=(1.0, 1.31), step_factor=(0.01, 0.03))
    ],
    
)
```

# 03 generate defect masks with randomly applied augmentations

```python
z = 0

desired_num_of_masks = 2300

for j, f in enumerate(list(os.listdir("./defect_masks/"))):
    full_f = os.path.join("./defect_masks/", f)
    if not f.startswith("."):
        print(f)
        image = cv2.imread(full_f)

        y = 0
        while y < 100:
            print(j*100+y,"/",desired_num_of_masks)
            new_f = os.path.join("./defect_masks_gen/", "mask_{}.png".format(str(z).zfill(4)))
            transformed = transform(image=image)
            if transformed['image'].max() > 127:
                cv2.imwrite(new_f, transformed['image'])
                y += 1
                z += 1
```
